```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data= pd.read_excel('/content/drive/MyDrive/496/Online Retail.xlsx')
data.head()
```

```python
data.columns
```

```python
data['Description']=data['Description'].str.strip()
data.head()

data.dropna(axis= 0, subset = ['InvoiceNo'], inplace = True)
data['InvoiceNo']=data['InvoiceNo'].astype('str')

data=data[~data['InvoiceNo'].str.contains('C')]
```

```python
basket_Italy = (data[data['Country']=="Italy"]
              .groupby(['InvoiceNo','Description'])['Quantity']
              .sum().unstack().reset_index().fillna(0)
              .set_index('InvoiceNo'))
print(basket_Italy.head())
```

```python
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
# Encoding the datasets
basket_encoded = basket_Italy.applymap(hot_encode)
basket_Italy = basket_encoded
print(basket_Italy)
```

```python
frq_items = apriori(basket_Italy,min_support=0.05,use_colnames = True)
print(frq_items)
```

```python
rules = association_rules(frq_items, metric="lift", min_threshold=1)
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
print(rules.head(1))
```
