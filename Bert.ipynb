```python
import torch
from transformers import BertTokenizer, BertForMaskedLM

```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')
```

```python
sequence = "He began his premiership by forming a five-man war cabinet which included Chamberlain as Lord President"
inputs = tokenizer.encode_plus(sequence, return_tensors='pt', padding=True, truncation=True)
outputs = model(**inputs)
predictions = outputs.logits.argmax(dim=-1)
prediction_text = tokenizer.decode(predictions[0], skip_special_tokens=True)
```

```python
print(prediction_text)
```
