```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

#initialize tokenize and model from pretrained GPT2 model
tokenizer= GPT2Tokenizer.from_pretrained('gpt2')
model=GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
#sequence : your words, doesn't need to follow this
sequence='He began his premiership by forming a five-man war cabinet which included chamerlain as Lord President '
inputs= tokenizer.encode(sequence, return_tensors='pt')
outputs=model.generate(inputs, max_length=200, do_sample=True, temperature=1)
tokenizer.decode(outputs[0], skip_special_tokens=True)
```
