```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
#reading data from csv and putting them in pandas sataframe
titanicTrain=pd.read_csv('/content/drive/MyDrive/496/cleanTitanicTrain.csv')
titanicTest=pd.read_csv('/content/drive/MyDrive/496/cleanTitanicTest.csv')
```

```python
#pick the predators among all possible feature, we have 12, we pick 7
predictorNames=["Pclass", "Sex", "Age","SibSp","Parch", "Fare", "Embarked"]
trainPredictors=titanicTrain[predictorNames]
#create a list with a single value: survived, not survived
targetName=['Survived']
trainTarget=titanicTrain['Survived']
#you do the same to pick 7 of 12 features from the test dataset
testPredictors=titanicTest[["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
```
